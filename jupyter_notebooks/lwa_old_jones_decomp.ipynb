```python
import numpy as np

from pyuvdata import UVBeam

filename = "/Users/bryna/Projects/Physics/data_files/lwa_beam_testing/LWAbeam_2015.fits"
lwa_beam = UVBeam.from_file(filename, file_type = "beamfits")
```

```python
lwa_beam.select(axis2_inds=np.nonzero(lwa_beam.axis2_array <= np.pi/2))
```

```python
lwa_beam.plot(freq_ind=-1, complex_type="abs")
```

```python
lwa_beam.plot(freq_ind=-1, complex_type="real")
```

```python
lwa_beam.plot(freq_ind=-1, complex_type="imag")
```

```python
lwa_beam.plot(freq_ind=-1, complex_type="phase")
```
